```python
df['SMOKING']=df['SMOKING'].map({1:'NO',2:'YES'})
```

```python
# Smokers count
plt.figure(figsize=(10,10))
sns.countplot(x="SMOKING", data=df, palette="plasma")
plt.xlabel('SMOKING')
plt.ylabel('Count')
plt.show()
```

```python
# Percentage of people with lung cancer in the dataset
plt.figure(figsize = (13, 13))
df.LUNG_CANCER.value_counts().plot(kind='pie',figsize=(8, 8),autopct='%1.1f%%')
plt.show()
```

```python
# Lung Cancer across Genders
plt.figure(figsize=(10,10))
sns.countplot(data=df,x='GENDER',hue='LUNG_CANCER',palette='plasma')
plt.xlabel('GENDER')
plt.legend(["Has cancer", 'Does not have cancer'])
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df,x='COUGHING',hue='LUNG_CANCER')
plt.legend(["Has cancer", 'Does not have cancer'])
```
